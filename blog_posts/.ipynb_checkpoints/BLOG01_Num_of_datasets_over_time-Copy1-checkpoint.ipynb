```python
def convert_file_size(file_size):
    '''
    Utility function to convert units of file size
    '''
    unit_dict = {
        1: 'B',
        2: 'KB',
        3: 'MB',
        4: 'GB',
        5: 'TB',
        6: 'PB'
    }
    
    if file_size > 0:
        i = 0
        while file_size >= (1024 ** i):
            i += 1
        unit = unit_dict[i]
        new_size = file_size / (1024 ** (i-1))
        size_string = ('%.1f' % new_size) + unit
    else:
        size_string = '0B'
        
    return size_string
```

```python
def make_category_dummies(series, unique_categories):
    '''
    For a given pd.Series and a set of categories from a categorical variable, create dummy vriables
    and count frequency of each category in the series.
    '''
    df = pd.DataFrame()
    for category in unique_categories:
        df[category] = (series == category)
    
    return df.astype(int)

def make_category_table(df, series, unique_categories):
    dummies = make_category_dummies(series, unique_categories)
    table = pd.concat([df['labHead_1 name'], dummies], axis=1).groupby('labHead_1 name').sum()
    return table
```

# Import Data

```python
# Importing the DB, append 'Journal' from pubmed-scraped DB
journals = pd.read_csv('../data/scraped_pubmed_latest.csv', usecols=['Dataset ID', 'Journal'])

px = pd.read_csv('../data/scraped_xml_latest.csv', low_memory=False)
px = px.merge(journals, on='Dataset ID', how='left')

pride = pd.read_csv('../data/scraped_pride_latest.csv', low_memory=False)
pride = pride.merge(journals, on='Dataset ID', how='left')

# Remove journals variable from memory
del journals
gc.collect()
```

```python
# Check initial lengths of the two DBs before processing
print('# of datasets in PX-XML DB: %d' % len(px))
print('# of datasets in PRIDE-JSON DB: %d' % len(pride))
```

```python
# Convert submissionDate to datetime object
px['Date'] = px['Announce Date'].str.extract(r'(201[0-9]-[0-9]{2}-[0-9]{2})')
px = px[px['Date'].notnull()]
px.loc[:, 'Date'] = pd.to_datetime(px['Date'], infer_datetime_format=True)
px['year'] = px['Date'].dt.year
px['month'] = px['Date'].dt.month
del px['Announce Date']

pride.loc[:, 'submissionDate'] = pd.to_datetime(pride['submissionDate'])#, infer_datetime_format=True)
pride = pride[pride['submissionDate'].notnull()]
pride['year'] = pride['submissionDate'].dt.year
pride['month'] = pride['submissionDate'].dt.month

# Drop Day Part, convert back to datetime object. This will make the 'day' part all 01 so months can be tracked
px['submissionDate'] = px['year'].map(str) + '-' + px['month'].map(str)
px.loc[:, 'submissionDate'] = pd.to_datetime(px['submissionDate'])

pride['submissionDate2'] = pride['year'].map(str) + '-' + pride['month'].map(str)
pride.loc[:, 'submissionDate2'] = pd.to_datetime(pride['submissionDate2'])
```

```python
tally = px_ftp.groupby(['Dataset ID', 'HostingRepository', 'submissionDate']).sum().reset_index()
tally['size'] = tally['size_actual'].apply(lambda x: convert_file_size(x))
```

```python
tally.head()
```

```python
separate2 = tally[['submissionDate', 'HostingRepository', 'size_actual']]\
                            .groupby(['submissionDate', 'HostingRepository']).sum().unstack('HostingRepository')
separate2 = separate2.replace({np.NaN: 0})
separate2.columns = separate2.columns.droplevel()
separate2['total'] = separate2.sum(axis=1)
separate['date_str'] = separate.index.strftime('%Y-%m')
```

```python
# 1TB = 1024^4 B
separate2 = separate2 / (1024 ** 4)
```

```python
sep2 = separate2[['PRIDE', 'MassIVE']].replace({np.NaN:0})
sep2['counts'] = sep2.values.tolist()
sep2.head(10)
```

# Word Cloud each Repository

```python
from wordcloud import WordCloud
from wordcloud import WordCloud, STOPWORDS
```

```python
stopwords = set(STOPWORDS)
```

```python
repos_labels = ['PRIDE', 'MassIVE', 'PeptideAtlas', 'iProX', 'jPOST']
m = len(repos_labels)
fig = plt.figure(figsize=(20, 15))
for i, label in enumerate(repos_labels):
    d = px[px.HostingRepository == label][['Description', 'Title', 'curator keywords', 'journal article keyword', 'submitter keywords']]
    text = d.replace({np.NaN:''}).sum().sum()
    text = text.replace('protein', '')
    text = text.replace('proteome', '')
    text = text.replace('proteomic', '')
    text = text.replace('analysis', '')
    text = text.replace('identified', '')
    text = text.replace('set', '')
    text = text.replace('data', '')
    text = text.replace('contains', '')
    text = text.replace('studies', '')
    text = text.replace('study', '')
    text = text.replace('cell', '')
    text = text.replace('using', '')
    text = text.replace('gene', '')
    text = text.replace('used', '')
    text = text.replace('et al', '')
    text = text.replace('sample', '')
    text = text.replace('including', '')
    text = text.replace('involved', '')
    wc = WordCloud(background_color="white", stopwords=stopwords).generate(text)
    
    ax = fig.add_subplot(3, 2, i+1)
    ax.set_title('Word Cloud for %s' % label, fontsize=16)
    ax.imshow(wc, cmap=plt.cm.gray, interpolation='bilinear')
    ax.axis('off')

fig.tight_layout()
```
