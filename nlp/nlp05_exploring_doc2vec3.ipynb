##### Doc2Vec

```python
nlp = spacy.load('en_core_web_lg')
```

```python
model3 = Doc2Vec.load('d2v_test3.model')
```

```python
def vectorize_abstract(model, abstract):
    doc = nlp(abstract)
    sentences = [sent.text for sent in doc.sents]
    vecs = []
    for s in sentences:
        v = model.infer_vector(s)
        vecs.append(v)
    return len(vecs), sum(vecs)
```

```python
%%time
abstracts = list(df['Abstract'])
abst_vectors_sum = []
abst_vectors_mean = []
for abst in abstracts:
    length, vecs_sum = vectorize_abstract(model3, abst)
    abst_vectors_sum.append(vecs_sum)
    abst_vectors_mean.append(vecs_sum/length)
```

```python
vecs_sum = np.array(abst_vectors_sum)
vecs_mean = np.array(abst_vectors_mean)
```
