# Vectorize with Doc2Vec model

```python
nlp = spacy.load('en_core_web_lg')
```

```python
model = Doc2Vec.load('models/d2v_descriptive.model')
```

```python
# Take a text consiting of multiple lines, vectorize each sentence
# and sum up the sentences. Returns the list of vectors as well as number of vectors.
def vectorize_text(model, text):
    doc = nlp(text)
    sentences = [sent.text for sent in doc.sents]
    vecs = []
    for s in sentences:
        v = model.infer_vector(s)
        vecs.append(v)
    return len(vecs), sum(vecs)
```

```python
%%time
vectors_sum = []
vectors_mean = []
for txt in descriptive_texts:
    length, vecs_sum = vectorize_text(model, txt)
    vectors_sum.append(vecs_sum)
    vectors_mean.append(vecs_sum/length)
```
