## Unlike nlp03 notebook, try using each sentence, instead of a whole abstract, as the 'doc'

```python
def process_abstract2(abstracts):
    sentences = []
    for abst in abstracts:
        doc = nlp(abst)
        s = [sent.text for sent in doc.sents]
        sentences.extend(s)
    return sentences
```

```python
process_abstract2(df.iloc[:1]['Abstract'])
```

```python
%%time
n = len(df)
abstracts = df.iloc[:n]['Abstract']
data = process_abstract2(abstracts)
# for i in range(n):
#     abst = abstracts2[i]
#     tokenized_doc = process_abstract(abst)
#     data2.append(tokenized_doc)
```

```python
tagged_data = [TaggedDocument(
                            words=word_tokenize(s.lower()), 
                            tags=[str(i)]) for i, s in enumerate(data)]
```

```python
%%time
# Train the model
max_epochs = 100
vec_size = 300
alpha = 0.025

model = Doc2Vec(
    vector_size=vec_size,
    alpha=alpha,
    min_alpha=0.025,
    min_count=1,
    dm=1,
    window=15,
    sample=0, #1e-5,
    negative=20, #5,
    workers=4
)

model.build_vocab(tagged_data)

for epoch in range(max_epochs):
    if epoch % 10 == 0:
        print('Iteraction {0}'.format(epoch))
    model.train(
        tagged_data,
        total_examples=model.corpus_count,
        epochs=model.epochs
    )
    
    # decrease the learning rate
    model.alpha -= 0.0002
    
    # fix the learning rate, no decay
    model.min_alpha = model.alpha

model.save('./d2v_test3.model')
print("Model Saved")
```

```python
model3 = Doc2Vec.load('d2v_test3.model')
```

```python
m = 2

similar_doc = model3.docvecs.most_similar(str(m))
print(similar_doc)

inds = [sim[0] for sim in similar_doc]

print('-'*5, 'Tag=%d' % m, '-'*10)
print(data[m])
for i in inds:
    print('-'*5, 'Tag=%s' % i, '-'*10)
    print(data[int(i)])
```

This particular example above seems not bad, but others I tried not so much. The model probably picks up the semantics better than the model trained with whole abstracts instead of sentences.

```python
def vectorize_abstract(model, abstract):
    doc = nlp(abstract)
    sentences = [sent.text for sent in doc.sents]
    vecs = []
    for s in sentences:
        v = model.infer_vector(s)
        vecs.append(v)
    return len(vecs), sum(vecs)
```

```python
%%time
abst_vectors_sum = []
abst_vectors_mean = []
for abst in abstracts:
    length, vecs_sum = vectorize_abstract(model3, abst)
    abst_vectors_sum.append(vecs_sum)
    abst_vectors_mean.append(vecs_sum/length)
```

```python
vecs_sum = np.array(abst_vectors_sum)
vecs_mean = np.array(abst_vectors_mean)
```
