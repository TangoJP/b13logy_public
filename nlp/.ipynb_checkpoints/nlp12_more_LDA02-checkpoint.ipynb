### Load lemmatized corpus

```python
infile = open('nlp09_data/nlp09_lemmatized_corpus.pickle', 'rb')
lemm_corpus = pickle.load(infile)
infile.close()
```

### Edit dictionary

```python
dictionary = corpora.Dictionary.load('nlp09_data/nlp09_lemmatized_corpus.dict')
```

```python
words2remove = [
    '', ' ', 'cell', 'study', 'project', 'approach', 'set', 
    '-pron-', 'show', 'data', 'dataset', 'file', 'biology', 'm.', 'n', 'reveal',
    'involve', 'thereby', 'affect', 'evaluation', 'elucidate', 'tool', 'read',
    'bp', 'upload', 'allow', 'investigate', 'find', 'know' 'insight', 'datum',
    'essential', 'many', 'well', 'lead', 'however', 'suggest', 'still', 'relate',
    'various', 'demonstrate', 'strongly', 'simple', 'notably', 'able',
    'furthermore', 'address', 'powerful', 'assume', 'respect', 'min', 'hour',
    'underlie', 'reflect', 'currently', 'largely', 'dedicated', 'explore', 'act',
    'play'
]
```

```python
remove_ids = [dictionary.token2id[word] for word in words2remove
                   if word in dictionary.token2id]
```

```python
dictionary.filter_tokens(remove_ids)
dictionary.compactify()
```

```python
dictionary.save('nlp12_data/nlp12_lemmatized_corpus.dict')
```

### Create a new BoW corpus

```python
# Create BoW
bow_corpus = [dictionary.doc2bow(text) for text in lemm_corpus]
corpora.MmCorpus.serialize('nlp12_data/nlp12_bow_corpus.mm', bow_corpus)
```

### Tf-Idf

```python
# Set up a tf-idf transforming model using the corpus
tfidf_model = models.TfidfModel(bow_corpus)

# corpus in tf-idf space
corpus_tfidf = tfidf_model[bow_corpus]
```

### LDA

```python
num_topics = 5
```

```python
# Transform tf-idf corpus with Latent Dirichilet Allocation
lda = models.LdaModel(
            corpus=corpus_tfidf, 
            id2word=dictionary, 
            num_topics=num_topics,
            alpha='auto'
)
```

```python
lda.save('nlp12_data/nlp12_tfidf_%d.lda' % num_topics)
```

```python
# corpus in LDA space
corpus_lda = lda[corpus_tfidf]
```

```python
print(tfidf_model)
print(len(corpus_tfidf))
print(len(corpus_lda))
```

```python
%%time
pyLDAvis.gensim.prepare(lda, corpus_tfidf, dictionary)
```

### Check how datasets spreads in the Topic Space

```python
def lda2df(corpus_lda):
    corpus_lda2_dict = []
    for doc in corpus_lda:
        lda2dict = {}
        for topic in doc:
            lda2dict[topic[0]] = topic[1]
        corpus_lda2_dict.append(lda2dict)
    df = pd.DataFrame(corpus_lda2_dict)
    return df.replace({np.NaN: 0})
```

```python
df = lda2df(corpus_lda)
```

```python
sns.set(style="darkgrid")
sns.pairplot(df.dropna(), plot_kws={'s': 5, 'color':'lightblue'})
```
